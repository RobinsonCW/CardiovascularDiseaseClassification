Let's create a function to find curve start and end dates and see if they match up with what Dr. Drew found. In Dr. Drew's notebook, "Curve day 1 starts when the daily new case count rises above a 0.00015 case to state population ratio."

The following function finds the start date for a curve. These are inline with Dr. Drew's dates, with the exception of Texas, having a curve start date of 6/23. Dr. Drew's notebook suggests 6/19.

The end dates don't match exactly either.

```python
def curve_start():
    states = ['NY','NJ','TX','CA']
    pops = [ny_pop, nj_pop, tx_pop, ca_pop]
    
    for state, pop in zip(states, pops):
        startFound = False
        
        cases = corona_set[corona_set['state'] == state][['date','positiveIncrease']].copy()
        cases.sort_values(by=['date'], inplace=True)
        
        print()
        
        for index, row in cases.iterrows():
            if startFound == False:
                if row['positiveIncrease'] / pop > .00015:
                    print(state, "Begin Date:", row['date'])
                    startFound = True
                    continue
                    
            if startFound == True:
                if row['positiveIncrease'] / pop < .00015:
                    print(state, "End Date:", row['date'])
                    break

curve_start()
```

## Modeling

Let's first introduce Dr. Drew's canned functions from his notebook.

https://nbviewer.jupyter.org/github/jakemdrew/CoronaCurves/blob/master/Corona%20Curve%20Forecast%20Texas%202.ipynb

```python
#Use mean absolute error (MAE) to score the regression models created 
#(the scale of MAE is identical to the response variable)
from sklearn.metrics import mean_absolute_error, make_scorer, mean_squared_error

#Function for Root mean squared error
#https://stackoverflow.com/questions/17197492/root-mean-square-error-in-python
def rmse(y_actual, y_predicted):
    return np.sqrt(mean_squared_error(y_actual, y_predicted))

#Function for Mean Absolute Percentage Error (MAPE) - Untested
#Adapted from - https://stackoverflow.com/questions/42250958/how-to-optimize-mape-code-in-python
def mape(y_actual, y_predicted): 
    mask = y_actual != 0
    return (np.fabs(y_actual - y_predicted)/y_actual)[mask].mean() * 100

#Create scorers for rmse and mape functions
mae_scorer = make_scorer(score_func=mean_absolute_error, greater_is_better=False)
rmse_scorer = make_scorer(score_func=rmse, greater_is_better=False)
mape_scorer = make_scorer(score_func=mape, greater_is_better=False)

#Make scorer array to pass into cross_validate() function for producing mutiple scores for each cv fold.
errorScoring = {'MAE':  mae_scorer, 
                'RMSE': rmse_scorer,
                'MAPE': mape_scorer
               } 
```

```python
from sklearn.model_selection import cross_validate

def EvaluateRegressionEstimator(regEstimator, X, y, cv):
    
    scores = cross_validate(regEstimator, X, y, scoring=errorScoring, cv=cv, return_train_score=True)

    #cross val score sign-flips the outputs of MAE
    # https://github.com/scikit-learn/scikit-learn/issues/2439
    scores['test_MAE'] = scores['test_MAE'] * -1
    scores['test_MAPE'] = scores['test_MAPE'] * -1
    scores['test_RMSE'] = scores['test_RMSE'] * -1

    #print mean MAE for all folds 
    maeAvg = scores['test_MAE'].mean()
    print_str = "The average MAE for all cv folds is: \t\t\t {maeAvg:.5}"
    print(print_str.format(maeAvg=maeAvg))

    #print mean test_MAPE for all folds
    scores['test_MAPE'] = scores['test_MAPE']
    mape_avg = scores['test_MAPE'].mean()
    print_str = "The average MAE percentage (MAPE) for all cv folds is: \t {mape_avg:.5}"
    print(print_str.format(mape_avg=mape_avg))

    #print mean MAE for all folds 
    RMSEavg = scores['test_RMSE'].mean()
    print_str = "The average RMSE for all cv folds is: \t\t\t {RMSEavg:.5}"
    print(print_str.format(RMSEavg=RMSEavg))
    print('*********************************************************')

    print('Cross Validation Fold Mean Error Scores')
    scoresResults = pd.DataFrame()
    scoresResults['MAE'] = scores['test_MAE']
    scoresResults['MAPE'] = scores['test_MAPE']
    scoresResults['RMSE'] = scores['test_RMSE']
    return scoresResults
```

### Option 1

In option 1, we'll use NY and CA data for the prediction. 

Let's build a simple polynomial regression model (degree = 4) using just curve day as input and cases as our response. We'll first create a filtered data set, to remove TX and CA data, along with anything prior to the curve start = 1. Also, we will remove everything past 7/6/2020 (per assignment).

```python
cut_off_date = dt.datetime(2020, 7, 6)

filtered_set = corona_set[corona_set['curve_day'] > 0].copy()
filtered_set = filtered_set[filtered_set['state'].isin(['NJ','NY'])]
filtered_set = filtered_set[filtered_set['date'] <= cut_off_date]

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))
sns.lineplot(x="curve_day", y="cases", data=filtered_set, hue="state", ax=ax1)
sns.lineplot(x="curve_day", y="tests", data=filtered_set, hue="state", ax=ax2)
plt.show()
```

```python
cols = ['curve_day']
#cols = ['curve_day','tests']
#cols = ['curve_day','tests','hospitalizedCurrently']

X = filtered_set[cols]
y = filtered_set['cases']

#If Single Feature, we need to reshape
#X = X.reshape(-1, 1)

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(4)
poly.fit(X)
X_poly = poly.transform(X)

#print(X_poly.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

# Using K-fold
cv = KFold(n_splits=10)

#Create a Linear Regression object and perform a grid search to find the best parameters
linreg = LinearRegression()
parameters = {'normalize':(True,False), 'fit_intercept':(True,False)}

#Create a grid search object using the  
from sklearn.model_selection import GridSearchCV
regGridSearch = GridSearchCV(estimator=linreg
                   , verbose=1 # low verbosity
                   , param_grid=parameters
                   , cv=cv # KFolds = 10
                   , scoring=mae_scorer)

#Perform hyperparameter search to find the best combination of parameters for our data
regGridSearch.fit(X_poly, y)
fit = linreg.fit(X_poly, y)
print(fit.score(X_poly, y))

clf = regGridSearch.best_estimator_
EvaluateRegressionEstimator(clf, X_poly, y, 10)
```

Average MAE during cross validation is 4.2692. We can convert this to cases using the following formula

**(MAE * Population) / 100,000**

For New York, this comes out to 803 cases.

Let's show the polynomial fit line against our observed data.

```python
yhat = clf.predict(X_poly)

#new_set['yhat'] = yhat
#pal = sns.dark_palette('yellow',2)

f, ax1 = plt.subplots(figsize=(16, 10))
sns.lineplot(x="curve_day", y="cases", data=filtered_set, hue="state", ax=ax1)
#sns.lineplot(x="x", y="tests", data=new_set, hue="state", ax=ax2)
sns.lineplot(x="curve_day", y=yhat, data=filtered_set, color='black')
plt.show()
```
